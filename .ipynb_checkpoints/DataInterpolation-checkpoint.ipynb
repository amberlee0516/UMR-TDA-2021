```python
def Distance(row1, row2):
    #print(row1.LATITUDE)
    #print(row1.LONGITUDE)
    #print(row2.LATITUDE)
    #print(row2.LONGITUDE)
    series1 =(row1.LATITUDE, row1.LONGITUDE)
    series2 = (row2.LATITUDE, row2.LONGITUDE)
    dist = distance.distance(series1, series2).km
    return dist

def fillMatrix(df):
    dist_matrix = np.zeros((df.shape[0],df.shape[0]))
#     row1Index = df.iloc[0]
#     #print(row1Index)
#     for index, row in df.iterrows():
#         dist_matrix[]
    size = dist_matrix.size
    #print("size = " + str(size))
    for i in range(0, df.shape[0]):
        #print("i = " + str(i))
        #print("\n")
        loc_i = df.iloc[i]
        for j in range(i+1, df.shape[0]):
            #print("j = " + str(j))
            loc_j = df.iloc[j]
            dist_matrix[i,j] = Distance(loc_i, loc_j)
            dist_matrix[j,i] = dist_matrix[i,j]
    return dist_matrix
    

    

```

```python
print(dummyData[0:1])
print(dummyData.shape)
print(dummyData)
row1 = pd.Series(dummyData.iloc[0])
row2 = pd.Series(dummyData.iloc[1])

print(Distance(row1, row2))
print("-------------")
print(dummyData.shape)
print(fillMatrix(dummyData))
print("Done")
```
