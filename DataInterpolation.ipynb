```python
def Distance(row1, row2):
    #print(row1.LATITUDE)
    #print(row1.LONGITUDE)
    #print(row2.LATITUDE)
    #print(row2.LONGITUDE)
    series1 =(row1.LATITUDE, row1.LONGITUDE)
    series2 = (row2.LATITUDE, row2.LONGITUDE)
    dist = distance.distance(series1, series2).km
    return dist

def fillMatrix(df):
    dist_matrix = np.zeros((df.shape[0],df.shape[0]))
#     row1Index = df.iloc[0]
#     #print(row1Index)
#     for index, row in df.iterrows():
#         dist_matrix[]
    size = dist_matrix.size
    #print("size = " + str(size))
    for i in range(0, df.shape[0]):
        #print("i = " + str(i))
        #print("\n")
        loc_i = df.iloc[i]
        for j in range(i+1, df.shape[0]):
            #print("j = " + str(j))
            loc_j = df.iloc[j]
            dist_matrix[i,j] = Distance(loc_i, loc_j)
            dist_matrix[j,i] = dist_matrix[i,j]
    return dist_matrix
 
    
def getMinimums(index, dist_matrix, df, numberOfMins, naIndices):
#     minimums = [0 for x in range(numberOfMins)]
#     mins_calculated = 0
    naIndices = np.ndarray.flatten(naIndices)
    temp = dist_matrix[index]
    print("this is the row of the distance matrix")
    print(temp)
    print("--------")
    temp = np.delete(temp, naIndices)
    print("The unsorted row of the distance matrix without the NA Values from the data")
    print(temp)
    print("--------")
    temp = np.sort(temp)
    temp = temp[temp != 0] # gets rid of zero distances
    distancesDesired =  temp[0:numberOfMins]

    goalIndices = []
    for x in range(0 , numberOfMins):
        goalIndices.append(np.where(dist_matrix[index] == distancesDesired[x]))
#     print("The goal indices to get TP values")
#     print(goalIndices)
#     print("The distances")
#     print(distancesDesired)
#     print("-----")
    return [distancesDesired, goalIndices]
    
    
def interpolate(n, minimums, df, naVar):
    result = 0
    denominator = sum(1 / minimums[0])
    distances = minimums[0]
    indices = minimums[1]
    
    for x in range(0, n):
        #print(df[naVar][minimums[1][x][0][0]])
        result +=  (1/distances[x]) / denominator * df[naVar][minimums[1][x][0][0]] 
#     print("The interpolated value:")
#     print(result)
#     print("-------")
    return result

        
    
# Predicts using linear / kth nearest points of the same value     
def predict(df, dist_matrix, naVar, n):
    df["PREDICTED_" + naVar] = df[naVar]
    naIndices = np.argwhere(np.isnan(np.array(df[naVar])))
    for index in naIndices:
#         print("Index in Quesiton:")
#         print(index)
#         print("--------")
        minimums = getMinimums(index[0], dist_matrix, df, n, naIndices) # get n min distances
        print("Result of Calling Minimums")
        print(minimums)
        print("--------")
#         print(minimums)
#         print("The value that is about to get changed")
#         print(df["PREDICTED_" + naVar][index])
#         print("-------")
        predictedVal = interpolate(n, minimums, df, naVar)
        df["PREDICTED_" + naVar][index[0]] = predictedVal
    return df
    

    
    
def filterUniqueSheetbars(df):
    result = pd.DataFrame().reindex_like(df)
    duplicateSheetbars = pd.DataFrame(df.groupby("SHEETBAR").size(),
                                     columns = ['count']).reset_index()
    print(duplicateSheetbars)
    
    return df

print("done")
```
